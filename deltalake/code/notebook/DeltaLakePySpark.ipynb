**Create a table**

```python
data = spark.range(0, 5)
#display(data)

data.write.mode("append").format("delta").option("mergeSchema", "true").save("Tables/deltademo")
```

**Read data**

```python
# Orginal Data

df = spark.read.format("delta").load("Tables/deltademo")
df.show()
```

**Update table data**

```python
# Overwrite

data = spark.range(5, 10)
data.write.format("delta").mode("overwrite").save("Tables/deltademo")
```

```python
# Overwriten Data

df = spark.read.format("delta").load("Tables/deltademo")
df.show()
```

```python
# Append

data = spark.range(11, 15)
data.write.format("delta").mode("append").save("Tables/deltademo")
```

```python
# Red Appended Data

df = spark.read.format("delta").load("Tables/deltademo")
df.show()
```

**Read older versions of data using time travel**

```python
# Time Travel

df = spark.read.format("delta") \
  .option("versionAsOf", 0) \
  .load("Tables/deltademo")

df.show()
```

```python
# Time Travel2

dftt2 = spark.read.format("delta") \
  .option("versionAsOf", 2) \
  .load("Tables/deltademo")

dftt2.show()
```

```python
# Read Diabetes

dfdiabetes = spark.read.format("csv").option("header","true").load("Files/diabetes/diabetes.csv")
display(dfdiabetes)
```

```python
# Write a table
dfdiabetes.write.mode("append").format("delta").option("mergeSchema", "true").save("Tables/deltadiabetes")
```

```python
%%sql

select count(*) from deltadiabetes
```

```python
# Read New Diabetes

dfnew = spark.read.format("csv").option("header","true").load("Files/diabetes/newdiabetes.csv")
display(dfnew)
```

```python
# Write new data to Delta table
dfnew.write.mode("append").format("delta").option("mergeSchema", "true").save("Tables/deltadiabetes")
```

```python
%%sql

select count(*) from deltadiabetes
```

```python
# Write new data to Delta table
dfnew.write.mode("overwrite").format("delta").option("mergeSchema", "true").save("Tables/deltadiabetes")
```
