```python
from pyspark.sql.types import *
from  pyspark.sql.functions import *

def loadFullDataFromSource(file_name):

    df = spark.sql("SELECT " + file_name + " as scalekey, fourdigityear as year, drugkey, geokey, Prscrbr_NPI, Tot_Clms, Tot_30day_Fills, Tot_Day_Suply, Tot_Drug_Cst, Tot_Benes, GE65_Bene_Sprsn_Flag, GE65_Tot_Clms, GE65_Tot_30day_Fills, GE65_Tot_Drug_Cst, GE65_Tot_Day_Suply, GE65_Sprsn_Flag, GE65_Tot_Benes FROM medicarepartd.medicarepartdwithkeys")

    df.write.mode("append").format("delta").option("mergeSchema", "true").option("schema", "schematable").save("Tables/factmedicarepartd")

full_files = [
    '1',
    '2',
    '3',
    '4',
    '5'
]

for file in full_files:
    loadFullDataFromSource(file)
```
